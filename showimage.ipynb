```python
import numpy as np
import cv2
import pandas as pd
```

```python
data = pd.read_csv('test.csv')
```

```python
data.head(10)
```

```python
cv2.namedWindow('Images', cv2.WINDOW_NORMAL)

num_rows = 6
num_cols = 5
window_width = 28 * num_cols
window_height = 28 * num_rows
cv2.resizeWindow('Images', window_width, window_height)

images = []

for i in range(30):
    one = data.loc[i:i, 'pixel0':]
    pixel_string = one.apply(lambda x: ','.join(x.astype(str)), axis=1)
    pixels = pixel_string.str.split(',').apply(lambda x: [int(i) for i in x]).tolist()
    pixels = np.array(pixels, dtype=np.uint8)

    width = 28
    height = 28
    channels = 1
    image = pixels.reshape((height, width, channels))
    images.append(image)

# Xây dựng lưới ảnh
grid = np.zeros((window_height, window_width), dtype=np.uint8)
for i in range(num_rows):
    for j in range(num_cols):
        index = i * num_cols + j
        if index < len(images):
            y_start = i * height
            y_end = (i + 1) * height
            x_start = j * width
            x_end = (j + 1) * width
            grid[y_start:y_end, x_start:x_end] = images[index].reshape((height, width))

cv2.imshow('Images', grid)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
